```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import os
```

```python
data = pd.read_csv("sales_data.csv",encoding='latin1')
```

```python
data.dropna(subset=['STATE'],inplace = True)
```

```python
data.dropna(subset = ['TERRITORY'],inplace = True)
```

```python
data['ADDRESSLINE2'] = data['ADDRESSLINE2'].fillna('0')
```

```python
data['POSTALCODE'].fillna(0)
```

**EXPLORING DATA** 

```python
data.columns
```

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
data.describe()
```

```python
monthly_sales = data.groupby(data['ORDERDATE'].dt.to_period('M'))['SALES'].sum().reset_index()
```

```python
monthly_sales['ORDERDATE'] = monthly_sales['ORDERDATE'].dt.to_timestamp()
```

**plot**

```python
plt.figure(figsize = (12,6))
sns.lineplot(x = 'ORDERDATE',y = 'SALES',data = monthly_sales)
plt.title('MOnthly Sales over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.show()
```

**SALES BY PRODUCT LINE**

```python
sales_by_product = data.groupby('PRODUCTLINE')['SALES'].sum().reset_index()
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = 'PRODUCTLINE',y = 'SALES' , data = sales_by_product)
plt.title('Sales by Product line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

**SALES BY COUNTRY**

```python
sales_by_country = data.groupby('COUNTRY')['SALES'].sum().reset_index()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'COUNTRY', y = 'SALES' , data = sales_by_country)
plt.title('Sales by country')
plt.xlabel('Country')
plt.ylabel('Total sales')
plt.xticks(rotation = 90)
plt.show()
```

**SALES BY STATUS**

```python
sales_by_status = data.groupby('STATUS')['SALES'].sum().reset_index()
```

```python
sns.barplot(data = sales_by_status, x='STATUS',y='SALES')
```

**CORRELATION MATRIX**

```python
numerical_data = data.select_dtypes(include=[np.number])
corr_matrix = numerical_data.corr()
plt.figure(figsize = (10,8))
sns.heatmap(corr_matrix,annot = True,cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**MARKDOWN BY SALES**

```python
plt.figure(figsize = (8,6))
sns.histplot(data['SALES'],kde = True)
plt.title('Distribution by sales')
```

**SALES BY QUARTER**

```python
sales_by_quarter = data.groupby('QTR_ID')['SALES'].sum().reset_index()
plt.figure(figsize = (8,6))
sns.barplot(x = 'QTR_ID',y='SALES',data = sales_by_quarter)
```

**SALES BY MONTH**

```python
sales_by_month = data.groupby('MONTH_ID')['SALES'].sum().reset_index()
sns.lineplot(data = sales_by_month,x='MONTH_ID',y='SALES')
```

```python
sns.barplot(data = sales_by_month,x='MONTH_ID',y='SALES')
```

**SALES BY YEAR**

```python
sales_by_year = data.groupby('YEAR_ID')['SALES'].sum().reset_index()
```

**TOP CUSTOMERS BY SALES**

```python
top_customers = data.groupby('CUSTOMERNAME')['SALES'].sum().reset_index().sort_values(by = 'SALES',ascending = False).head(15)
sns.barplot(x='SALES',y= 'CUSTOMERNAME',data = top_customers)                                                                                      
```

```python
sales_by_deal_size = data.groupby('DEALSIZE')['SALES'].sum().reset_index()
sns.barplot(data = sales_by_deal_size,x='DEALSIZE',y='SALES')
```
